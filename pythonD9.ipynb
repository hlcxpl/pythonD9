```python
import pandas as pd

# Cargar datos desde el archivo CSV
data = pd.read_csv('worldcup2014.csv')

# Muestra el DataFrame para verificar si los datos se cargaron correctamente
data
```

```python


# a. Pertenecer a África
prob_africa = len(data[data['continent'] == 'africa']) / len(data)

# b. Pertenecer a Asia
prob_asia = len(data[data['continent'] == 'asia']) / len(data)

# c. Pertenecer a Europa
prob_europa = len(data[data['continent'] == 'europe']) / len(data)

# d. Pertenecer a Norteamérica
prob_norteamerica = len(data[data['continent'] == 'northamerica']) / len(data)

# e. Pertenecer a Sudamérica
prob_sudamerica = len(data[data['continent'] == 'southamerica']) / len(data)

# f. Clasificar a segunda ronda
prob_clasificar_segunda_ronda = len(data[data['group_pos'] <= 2]) / len(data)

# g. Ganar al menos un partido
prob_ganar_al_menos_un_partido = len(data[data['juegos_ganados'] >= 1]) / len(data)

# h. Pertenecer a África y clasificar a segunda ronda
prob_africa_y_clasificar_segunda_ronda = len(data[(data['continent'] == 'africa') & (data['group_pos'] <= 2)]) / len(data)

# i. Pertenecer a Europa y clasificar a segunda ronda
prob_europa_y_clasificar_segunda_ronda = len(data[(data['continent'] == 'europe') & (data['group_pos'] <= 2)]) / len(data)

# j. Pertenecer a Asia y clasificar a segunda ronda
prob_asia_y_clasificar_segunda_ronda = len(data[(data['continent'] == 'asia') & (data['group_pos'] <= 2)]) / len(data)

# k. Pertenecer a Sudamérica y clasificar a segunda ronda
prob_sudamerica_y_clasificar_segunda_ronda = len(data[(data['continent'] == 'southamerica') & (data['group_pos'] <= 2)]) / len(data)

# l. Pertenecer a Norteamérica y clasificar a segunda ronda
prob_norteamerica_y_clasificar_segunda_ronda = len(data[(data['continent'] == 'northamerica') & (data['group_pos'] <= 2)]) / len(data)

# m. Pertenecer a África o clasificar a segunda ronda
prob_africa_o_clasificar_segunda_ronda = len(data[(data['continent'] == 'africa') | (data['group_pos'] <= 2)]) / len(data)

# n. Pertenecer a Europa o clasificar a segunda ronda
prob_europa_o_clasificar_segunda_ronda = len(data[(data['continent'] == 'europe') | (data['group_pos'] <= 2)]) / len(data)

# o. Pertenecer a Asia o clasificar a segunda ronda
prob_asia_o_clasificar_segunda_ronda = len(data[(data['continent'] == 'asia') | (data['group_pos'] <= 2)]) / len(data)

# p. Pertenecer a Sudamérica o clasificar a segunda ronda
prob_sudamerica_o_clasificar_segunda_ronda = len(data[(data['continent'] == 'southamerica') | (data['group_pos'] <= 2)]) / len(data)

# q. Pertenecer a Norteamérica o clasificar a segunda ronda
prob_norteamerica_o_clasificar_segunda_ronda = len(data[(data['continent'] == 'northamerica') | (data['group_pos'] <= 2)]) / len(data)

# r. Pertenecer a África y haber ganado al menos un partido
prob_africa_y_ganar_al_menos_un_partido = len(data[(data['continent'] == 'africa') & (data['juegos_ganados'] >= 1)]) / len(data)

# s. Pertenecer a Europa y haber ganado al menos un partido
prob_europa_y_ganar_al_menos_un_partido = len(data[(data['continent'] == 'europe') & (data['juegos_ganados'] >= 1)]) / len(data)

# t. Pertenecer a Asia y haber ganado al menos un partido
prob_asia_y_ganar_al_menos_un_partido = len(data[(data['continent'] == 'asia') & (data['juegos_ganados'] >= 1)]) / len(data)

# u. Pertenecer a Sudamérica y haber ganado al menos un partido
prob_sudamerica_y_ganar_al_menos_un_partido = len(data[(data['continent'] == 'southamerica') & (data['juegos_ganados'] >= 1)]) / len(data)

# v. Pertenecer a Norteamérica y haber ganado al menos un partido
prob_norteamerica_y_ganar_al_menos_un_partido = len(data[(data['continent'] == 'northamerica') & (data['juegos_ganados'] >= 1)]) / len(data)

# w. Haber ganado al menos un partido y clasificar a segunda ronda
prob_ganar_al_menos_un_partido_y_clasificar_segunda_ronda = len(data[(data['juegos_ganados'] >= 1) & (data['group_pos'] <= 2)]) / len(data)

# Imprimir probabilidades
print("a. Pertenecer a África:", prob_africa)
print("b. Pertenecer a Asia:", prob_asia)
print("c. Pertenecer a Europa:", prob_europa)
print("d. Pertenecer a Norteamérica:", prob_norteamerica)
print("e. Pertenecer a Sudamérica:", prob_sudamerica)
print("f. Clasificar a segunda ronda:", prob_clasificar_segunda_ronda)
print("g. Ganar al menos un partido:", prob_ganar_al_menos_un_partido)
print("h. Pertenecer a África y clasificar a segunda ronda:", prob_africa_y_clasificar_segunda_ronda)
print("i. Pertenecer a Europa y clasificar a segunda ronda:", prob_europa_y_clasificar_segunda_ronda)
print("j. Pertenecer a Asia y clasificar a segunda ronda:", prob_asia_y_clasificar_segunda_ronda)
print("k. Pertenecer a Sudamérica y clasificar a segunda ronda:", prob_sudamerica_y_clasificar_segunda_ronda)
print("l. Pertenecer a Norteamérica y clasificar a segunda ronda:", prob_norteamerica_y_clasificar_segunda_ronda)
print("m. Pertenecer a África o clasificar a segunda ronda:", prob_africa_o_clasificar_segunda_ronda)
print("n. Pertenecer a Europa o clasificar a segunda ronda:", prob_europa_o_clasificar_segunda_ronda)
print("o. Pertenecer a Asia o clasificar a segunda ronda:", prob_asia_o_clasificar_segunda_ronda)
print("p. Pertenecer a Sudamérica o clasificar a segunda ronda:", prob_sudamerica_o_clasificar_segunda_ronda)
print("q. Pertenecer a Norteamérica o clasificar a segunda ronda:", prob_norteamerica_o_clasificar_segunda_ronda)
print("r. Pertenecer a África y haber ganado al menos un partido:", prob_africa_y_ganar_al_menos_un_partido)
print("s. Pertenecer a Europa y haber ganado al menos un partido:", prob_europa_y_ganar_al_menos_un_partido)
print("t. Pertenecer a Asia y haber ganado al menos un partido:", prob_asia_y_ganar_al_menos_un_partido)
print("u. Pertenecer a Sudamérica y haber ganado al menos un partido:", prob_sudamerica_y_ganar_al_menos_un_partido)
print("v. Pertenecer a Norteamérica y haber ganado al menos un partido:", prob_norteamerica_y_ganar_al_menos_un_partido)
print("w. Haber ganado al menos un partido y clasificar a segunda ronda:", prob_ganar_al_menos_un_partido_y_clasificar_segunda_ronda)
```

```python
# Probabilidades condicionales
prob_europa_condicional_clasificado = prob_europa_y_clasificar_segunda_ronda / prob_clasificar_segunda_ronda
prob_sudamerica_condicional_clasificado = prob_sudamerica_y_clasificar_segunda_ronda / prob_clasificar_segunda_ronda
prob_africa_condicional_gano_al_menos_un_partido = prob_africa_y_ganar_al_menos_un_partido / prob_ganar_al_menos_un_partido
prob_ganar_al_menos_un_partido_condicional_clasificado = prob_ganar_al_menos_un_partido_y_clasificar_segunda_ronda / prob_clasificar_segunda_ronda
prob_ganar_al_menos_un_partido_condicional_norteamerica = prob_norteamerica_y_ganar_al_menos_un_partido / prob_norteamerica
prob_no_clasificado_condicional_africa = (1 - prob_africa_y_clasificar_segunda_ronda) / (1 - prob_africa)

# Imprimir los resultados
print("a. Pertenecer a Europa, si se sabe que clasificó:", prob_europa_condicional_clasificado)
print("b. Pertenecer a Sudamérica, si se sabe que clasificó:", prob_sudamerica_condicional_clasificado)
print("c. Pertenecer a África, si se sabe que ganó al menos un partido:", prob_africa_condicional_gano_al_menos_un_partido)
print("d. Haber ganado al menos un partido, si se sabe que clasificó:", prob_ganar_al_menos_un_partido_condicional_clasificado)
print("e. Haber ganado al menos un partido, si se sabe que pertenece a Norteamérica:", prob_ganar_al_menos_un_partido_condicional_norteamerica)
print("f. NO haber clasificado, si se sabe que pertenece a África:", prob_no_clasificado_condicional_africa)

```
